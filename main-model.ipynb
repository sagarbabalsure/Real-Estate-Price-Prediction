```python
data2 = data.drop(['area_type','availability','society'],axis='columns')
data2.head()
```

```python
data2['balcony'].fillna(data2.balcony.mean(),inplace=True)
```

```python
data2['bath'].fillna(float(2),inplace=True)
```

```python
data3 = data2.dropna()
data3.isnull().sum()
```

```python
data3['size'].unique()
```

```python
data3['bhk'] = data3['size'].apply(lambda x : int(x.split(' ')[0]))
data3.head()
```

```python
location_stats = data3['location'].value_counts().sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_lthan_10 = location_stats[location_stats<=10]
location_stats_lthan_10
```

```python
data3['location'] = data3['location'].apply(lambda x: 'Other' if x in location_stats_lthan_10 else x)
data3
```

```python
data4 = data3.copy()
data4.head(10)
```

```python
data4.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data4[~data4['total_sqft'].apply(is_float)]
```

```python
def convert_to_float(x):
    token = x.split('-')
    if len(token) == 2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_to_float('2100 - 2850')
```

```python
data4['total_sqft'] = data4['total_sqft'].apply(convert_to_float)
data4.head()
```

```python
data4.total_sqft.unique()
```

```python
data4['price_per_sqft'] = (data4['price']*100000)/data4['total_sqft']
data4.head()
```

```python
data4[data4['total_sqft']/data4['bhk']<300].head()
```

```python
data5 = data4[~(data4.total_sqft/data4.bhk<300)]
data5.head()
```

```python
data6 = data5.drop(['size','price_per_sqft'],axis='columns')
data6.head()
```
