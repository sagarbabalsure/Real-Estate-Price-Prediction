## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Load 'Bengaluru_House_Data.csv' file

```python
data = pd.read_csv("Bengaluru_House_Data.csv")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data['area_type'].value_counts()
```

## Removing some unessential columns.

```python
data2 = data.drop(['area_type','availability','society','balcony'],axis='columns')
data2.head()
```

## Data cleaning.

```python
data2.isnull().sum()
```

```python
data2['bath'].value_counts()
```

```python
data2.isnull().sum()
```

```python
data3 = data2.dropna()
data3.isnull().sum()
```

```python
data3['size'].unique()
```

```python
data3['bhk'] = data3['size'].apply(lambda x : int(x.split(' ')[0]))
data3.head()
```

```python
location_stats = data3['location'].value_counts().sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_lthan_10 = location_stats[location_stats<=10]
location_stats_lthan_10
```

```python
data3['location'] = data3['location'].apply(lambda x: 'Other' if x in location_stats_lthan_10 else x)
data3
```

```python
data4 = data3.copy()
data4.head(10)
```

```python
data4.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data4[~data4['total_sqft'].apply(is_float)]
```

```python
def convert_to_float(x):
    token = x.split('-')
    if len(token) == 2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_to_float('2100 - 2850')
```

```python
data4['total_sqft'] = data4['total_sqft'].apply(convert_to_float)
data4.head()
```

```python
data4.total_sqft.unique()
```

```python
data4.isnull().sum()
```

```python
data5 = data4.dropna()
data5.isnull().sum()
```

```python
data5['price_per_sqft'] = (data5['price']*100000)/data5['total_sqft']
data5.head()
```

```python
data5[data5['total_sqft']/data5['bhk']<300].head()
```

```python
data5 = data5[~(data5.total_sqft/data5.bhk<300)]
data5.head()
```

```python
data5.price_per_sqft.describe()
```

## Remove Outlires.

```python
def remove_outliers(data):
    new_data = pd.DataFrame()
    for key, subdf in data.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_data = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        new_data = pd.concat([new_data, reduced_data],ignore_index=True)
    return new_data

data6 = remove_outliers(data5)
```

```python
data6.shape
```

```python
data6 = data6.drop(['size','price_per_sqft'],axis='columns')
data6.head()
```

## One hot encoding.

```python
dummies = pd.get_dummies(data6.location)
dummies.head(10)
```

```python
data7 = pd.concat([data6,dummies.drop('Other',axis='columns')],axis='columns')
data7.head()
```

```python
data8 = data7.drop(['location'],axis='columns')
data8.head()
```

```python
X = data8.drop(['price'],axis='columns')
X
```

```python
y = data8.price
y
```

```python
X.isnull().sum().head()
```

## Split the data for training and testing.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

## Train the model using supervised linear regression algorithm

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(X_train,y_train)
lreg.score(X_test,y_test)
```

## Predict the price of flat.

```python
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns == location)[0][0]
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index>=0:
        x[loc_index] = 1
    return lreg.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar', 1000, 3, 2)
```

```python
predict_price('1st Phase JP Nagar', 1000, 1, 1)
```
